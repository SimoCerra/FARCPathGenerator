```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
```

```python
costmap_path = os.path.abspath("costmap_astar.png")
map_path = os.path.abspath("Grugliasco/mask.jpg")
total_path = os.path.abspath("Grugliasco/total_path.npy")
map_orig = cv2.imread(map_path)
costmap_astar = cv2.imread(costmap_path)
total_path = np.load(total_path)
```

```python
orig_obstacles_indexes = np.argwhere(map_orig==0)
costmap_obstacles_indexes = np.argwhere(costmap_astar==0)
```

```python
plt.figure(figsize=(12,12))
plt.plot(costmap_obstacles_indexes[:,0],costmap_obstacles_indexes[:,1],'.g')
plt.plot(orig_obstacles_indexes[:,0],orig_obstacles_indexes[:,1],'.k')
```
